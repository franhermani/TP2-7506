```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from time import time

%matplotlib inline
```

```python
modelo = pd.read_csv('csv_files/modelo_final.csv')
modelo['post_por_nivel'] = 0
```

```python
si = modelo['sepostulo'] == 1
no = modelo['sepostulo'] == 0
modelo_si = modelo[si]
modelo_no = modelo[no]
```

```python
modelo__si = modelo_si.groupby(['nivel_laboral','idpostulante'])['post_por_nivel'].size().reset_index()
modelo__si.head(10)
```

```python
modelo = pd.merge(modelo,modelo__si,on=['nivel_laboral','idpostulante'],how='left')
```
